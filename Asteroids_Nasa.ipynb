###### 'Close Approach Date' & 'Orbit Determination Date'
Segregating Close Approach Year & Month

```python
df['Close Approach Year']=df['Close Approach Date'].apply(lambda x:x.split('-')[0])
```

```python
df['Close Approach Month']=df['Close Approach Date'].apply(lambda x:x.split('-')[1])
```

```python
df['Orbit Determination Year']=df['Orbit Determination Date'].apply(lambda x:x.split('-')[0])
```

```python
df['Orbit Determination Month']=df['Close Approach Date'].apply(lambda x:x.split('-')[1])
```

```python
df.drop(['Close Approach Date','Orbit Determination Date'],axis=1,inplace=True)
```

###### X and y variables

```python
X=df.drop(['Hazardous'],axis=1)
```

```python
y=df['Hazardous']
```

```python
X=pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators=[55,64,80,95,100,150,200,250,500]
max_features=[40,35,30,25,15,20,10,5]
```

```python
param_grid={'n_estimators':n_estimators,'max_features':max_features}
```

```python
rfc=RandomForestClassifier()
```

```python
grid_rfc=GridSearchCV(rfc,param_grid)
```

```python
grid_rfc.fit(X_train,y_train)
```

```python
grid_rfc.best_params_
```

```python
test_pred=grid_rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,test_pred))
```

```python
confusion_matrix(y_test,test_pred)
```
